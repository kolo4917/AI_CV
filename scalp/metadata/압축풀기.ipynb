```python
import os
import zipfile
import shutil

def extract_and_move_zip_files(src_directory, dest_directory):
    print(f"Searching in: {src_directory}")  # 디렉토리 탐색 시작 확인
    for root, dirs, files in os.walk(src_directory):
        print(f"Checking directory: {root}")  # 현재 확인 중인 디렉토리
        for file in files:
            if file.endswith('zip.part0'):
                print(f"Found ZIP: {file} in {root}")  # 찾은 ZIP 파일
                zip_path = os.path.join(root, file)
                # ZIP 파일 압축 해제
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(root)
                    print(f"Extracted: {zip_path}")
                # 압축 해제된 파일들을 지정된 폴더로 이동
                for item in zip_ref.namelist():
                    source_file = os.path.join(root, item)
                    destination_file = os.path.join(dest_directory, item)
                    shutil.move(source_file, destination_file)
                    print(f"Moved: {source_file} to {destination_file}")

# 경로 설정
src_directory = '/content/drive/MyDrive/ai_test/anotation_json/77'
dest_directory = '/content/drive/MyDrive/ai_test/anotation_json'

extract_and_move_zip_files(src_directory, dest_directory)
```
